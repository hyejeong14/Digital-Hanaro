#함수

##format()
- {}.format(x)
- {:d] => 정수형 숫자 출력
- {:숫자d] => 지정한 숫자의 칸에 맞춰 출력
- {:0숫자d} => 지정한 숫자의 칸에 맞추고 빈칸은 0으로 출력

```python
#숫자를 문자열로 변환하는 함수
a = "{}".format(23)
print(a)
print(type(a))
```

```python
"유퀴즈 상금 {} 만원".format(100)
```

```python
"{} {} {}".format(1,2,3) #중괄호 개수와 input 개수가 같아야 함
```

```python
a = "{:d}".format(23)
b = "{:5d}".format(23)
c = "{:05d}".format(23)
print("a : {}\nb : {}\nc : {}".format(a,b,c))
```

```python
a = "{:+d}".format(23)
b = "{:+d}".format(-23)
c = "{: d}".format(23)
d = "{: d}".format(-23)
print("a : {}\nb : {}\nc : {}\nd : {}".format(a,b,c,d))
```

```python
a = "{:+5d}".format(23)
b = "{:+5d}".format(-23)
c = "{:=+5d}".format(23)
d = "{:=+5d}".format(-23)
e = "{:=+05d}".format(23)
f = "{:=+05d}".format(-23)
print("a : {}\nb : {}\nc : {}\nd : {}\ne : {}\nf : {}".format(a,b,c,d,e,f))
```

```python
a = "{:f}".format(23.45)
b = "{:15f}".format(23.45)
c = "{:+15f}".format(23.45)
d = "{:+015f}".format(23.45)

print("a : {}\nb : {}\nc : {}\nd : {}".format(a,b,c,d))
```

```python
a = "{:15.3f}".format(23.456)
b = "{:15.2f}".format(23.456)
c = "{:15.1f}".format(23.456)

print("a : {}\nb : {}\nc : {}".format(a,b,c))
```

###format() vs f-string
*format() 함수를 사용하는 것이 더 좋은 경우*<br>
1) 문자열이 많을 때<br>
2) 여러 값을 리스트 형태로 사용하는 경우

```python
#1)문자열이 많을 때
mc = "유재석"
reward = 100

print('''\
유퀴즈는 다양한 사람을 만나서 이야기하는 프로이다.
엠씨는 {}이고, 퀴즈를 맞추면 상금 {} 만원을 준다.\
'''.format(mc,reward))

print("\n")

mc = "유재석"
reward = 100

print(f'''\
유퀴즈는 다양한 사람을 만나서 이야기하는 프로이다.
엠씨는 {mc}이고, 퀴즈를 맞추면 상금 {reward} 만원을 준다.\
''')

#format 함수의 경우 변수 변경 한 괄호 안에서 해결 가능 f-string의 경우 전문을 이동하며 수정해야 함
```

```python
#2) 여러 값을 리스트 형태로 사용할 경우
data = ["강백호",17,"M","난 천재니까"]
print("""\
이름 : {}
나이 : {}
성별 : {}
좌우명 : {}\
""".format(*data)) # format함수는 전개연산자 적용 가능 

print("\n")

data = ["강백호",17,"M","난 천재니까"]
print(f"""\
이름 : {data[0]}
나이 : {data[1]}
성별 : {data[2]}
좌우명 : {data[3]}\
""") 

#하나하나 다 입력할거야? ;; 
```

```python
s = "slamdunk01"
s.find("dunk")
```

###rfind()
- 오른쪽에서 왼쪽으로 찾아서 첫 번째 위치를 반환

```python
s = "0slamdunkslam01"
print(s.rfind("0"),"\n",s.rfind("slam"))
```

```python
print(s.find("slam"),"\n")
print(s.find("0"))
```

###학점을 알려주는 프로그램.,

```python
#학점을 알려주는 프로그램
# A : 3.5 ~ 4.5
# B : 2.5 ~ 3.5
# C : 1.5 ~ 2.5
# D : 0 ~ 1.5
# F : 0

score = float(input( "학점을 입력하세요 > "))

if 3.5<=score<=4.5 :
  print("A학점")
elif 2.5<=score :
  print("B학점")
elif 1.5<=score :
  print("C학점")
elif 0<score :
  print("D학점")
else :
  print("F학점")
```

#pass 키워드

- raise NotImplementedError 로 대체 <br>
=> 아직 구현되지 않은 부분임을 알리기 위해 일부러 에러를 발생
