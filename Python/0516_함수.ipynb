```python
def printntimes(value, n):
  for i in range(n): #n은 rangeㅇ에 들어가니까 정수여야함
    print(value)
printntimes("df",3)
```

```python
def printntimes(n,*values):
  for i in range(n):
    for j in values:
      print(j)
  
printntimes(3,"안녕","파이썬","잼ㅅ")
```

```python
def printntimes(*values,n=3):
  for i in range(n):
    for j in values:
      print(j)
  
printntimes("안녕","파이썬","잼ㅅ")
```

#함수의 활용

##재귀함수

```python
def factorial(n):
  output = 1
  for i in range(1,n+1):
    output *= i
  return output


factorial(3)
```

```python
#재귀함수로 구현
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)
    #n=3
    #3*factorial(2)
    #3*2*factorial(1)
    #3*2*1*factorial(0)
    
factorial(3)
```

#피보나치 수열

```python
#피보나치 수열 = 앞의 두 숫자를 합한게 다음 숫자
def fibonacci(n):
  if n==1:
    return 1
  elif n==2:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)

fibonacci(5)
```

```python
print(fibonacci(40))
```

#키워드 global

```python
count = 0
def fibonacci(n):
  global count #함수 밖의 변수 참조를 위해 global 사용
  count += 1
  if n==1:
    return 1
  elif n==2:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)
  print(count)

fibonacci(10)
```

#메모화

```python
a=[[1,2,3],[4,[5,6]],7,[8,9]]
def flat(data):
  output=[]
  for i in data:
    if type(i)==list:
      output += flat(i)
    else:
      output.append(i)
  return output
flat(a)
```

#람다
- 함수의 매개변수에 사용하는 함수 = 콜백 함수

```python
def repeat_10(x):
  for i in range(10):
    x() 
def print_hi():
  print("hi")


repeat_10(print_hi)
```

```python
#map 함수
#콜백 함수 적용 결과 값

def sqr(x):
  return x**2

list_a=[1,2,3]
output_map = map(sqr,list_a)
print(output_map)
print(list(output_map))
```

```python
#filter 함수

def under_3(x):
  return x<3

list_a=[1,2,3]
output_filter = filter(under_3,list_a)

print(output_filter)
print(list(output_filter))
type(output_filter)
```

```python
#람다 함수 => 일회용 함수

list_a=[1,2,3]
#콜백함수 자리에 람다를 쓴당
#사용 하는 방법 => lambda 매개변수 : 리턴값(기능의 결과 값)


output_map = map(lambda x:x**2,list_a)
print(output_map)
print(list(output_map))
```

```python
list_a=[1,2,3]
output_filter = filter(lambda x: x<3,list_a)

print(output_filter)
print(list(output_filter))
type(output_filter)
```
