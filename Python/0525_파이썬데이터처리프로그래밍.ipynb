###인덱스

####딕셔너리를 시리즈로 변환하기

```python
#이름 = 인덱스 이름 / 라벨
test_dict={
    'a':1,
    'b':2,
    'c':3
}

sr = pd.Series(test_dict)
```

```python
#값 확인
print(sr.values)

#딕셔너리의 키 => 인덱스 이름
#딕셔너리의 값 => 시리즈의 값 
```

####리스트를 시리즈로 변환하기

```python
#정수형 인덱스
#리스트 생성
test_list=['hello',23,3.14,True]

print(test_list)
```

```python
#리스트를 시리즈로 변환
sr = pd.Series(test_list)
print(sr)
```

```python
#인덱스 확인
print("인덱스 : ", sr.index)
print("값 : ", sr.values)


#리스트의 인덱스 => 시리즈의 인덱스
#리스트의 요소 => 시리즈의 값
```

###원소 선택

```python
#인덱스 사용
#tuple 생성 (tuple : 리스트와 유사, 소괄호 사용 & 변경 불가)

test_tuple=('정대만',184, 73.5, True)
print(test_tuple)
```

```python
#tuple을 Series로 변환
sr = pd.Series(test_tuple,index=['이름','키','체중','불꽃남자'])
print(sr)
```

```python
#하나의 원소 선택하기
print(sr[0]) #이름
print(sr[1]) #키
print(sr[2]) #체중
print(sr[3]) #불꽃남자
```

```python
print(sr['이름']) #이름
print(sr['키']) #키
print(sr['체중']) #체중
print(sr['불꽃남자']) #불꽃남자
```

```python
#여러 개 원소 선택하기 = 리스트 형태로 입력하기

# 1) 정수형 인덱스로 선택
print(sr[[0,3]])
print(sr[[1,2]])

# 2) 라벨 인덱스로 선택
print(sr[['이름','체중']])
print(sr[['키','체중']])
```

```python
#범위를 지정하여 여러 개의 원소 선택 = 슬라이싱
#정수형 인덱스 사용할 때 끝 번호는 포함되지 않음
print(sr[0:2]) #  = print(sr[0:1+1]) 출력:0번, 1번
print(sr[1:4])

#라벨 인덱스는 마지막 인덱스의 값까지 출력
print(sr['이름':'체중'])
```

##DataFrame

```python
#라이브러리 호출
import pandas as pd

#리스트를 값으로 가지는 딕셔너리 생성
test_dict={
    'c1':[1,2,3],
    'c2':[4,5,6],
    'c3':[7,8,9],
    'c4':[10,11,12]
    }
  
print(test_dict)
```

```python
#딕셔너리를 데이터프레임으로 변환
df = pd.DataFrame(test_dict)
print(df)
print(type(df))
#딕셔너리의 키는 열 이름, 값은 열의 값으로 사용
#딕셔너리의 값 = 리스트의 요소 개수 = 행 인덱스
```

###행, 열 이름 바꾸기

```python
#행 인덱스와 열 이름 설정하여 데이터프레임 생성
df = pd.DataFrame([['정대만',19,184],
                   ['서태웅',17,187]],
                   index=['선수1','선수2'],          #행 인덱스
                   columns=['이름','나이','키'])     #열 이름

print(df)
```

```python
#행 인덱스 확인
print(df.index)

#열 이름 확인
print(df.columns)
```

```python
#행 인덱스, 열 이름 => 이름 변경
df.index=['학생1','학생2']
df.columns = ['성명','연령','신장']
print(df)
```

```python
#메소드 함수 = 클래스 안에 있는 함수
#행 인덱스, 열 이름 => 이름 변경
#rename 메소드 함수의 반환값은 새로운 데이터프레임
df_new=df.rename(index = {'학생1':'선수1','학생2':'선수2'})
print(df_new)
print()
print(df)
```

```python
#원본을 바로 변경
df.rename(index = {'학생1':'선수1','학생2':'선수2'},inplace=True)
print(df)
```

```python
#열 이름 바꾸기
df_new = df.rename(columns={'성명' : '이름',
                            '연령' : '나이',
                            '신장' : '키'})

print(df)
print()
print(df_new)
```

```python
df.rename(columns={'성명' : '이름',
                   '연령' : '나이',
                   '신장' : '키'},inplace=True)
print(df)
```

##인덱스 활용

```python
import pandas as pd

sd = {
    "이름":['정대만','강백호','채치수'],
    "키":[184,189,193],
    "나이":[19,17,19],
    "득점":[28,1,20]
}

print(sd)
```

```python
df = pd.DataFrame(sd)
print(df)
```

###특정 열을 행 인덱스로 설정

set_index()

```python
print(df,'\n')
df_new = df.set_index('이름')
print(df_new)
#원래 행 인덱스(정수인덱스)가 없어지고 이름 column이 인덱스로 전환 
```

```python
print(df_new,'\n')
df_new_2 = df.set_index('키')
print(df_new_2)
#키 column이 index가 되고 행 인덱스였던 이름이 다시 column으로 돌아옴
```

###행 인덱스 재배열

reindex()
