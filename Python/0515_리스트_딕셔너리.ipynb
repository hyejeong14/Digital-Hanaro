```python
character ={
    "name":"정대만",
    "number":14,
    "record" : {"3point":24,"rebound":1},
    "school" : ["무석중","북산고"] 
}
```

```python
for i in character:
  ch=character[i]
  if type(ch)==dict:
    for j in ch:
      print(ch[j])
  elif type(ch)==list:
    for k in ch:
      print(k)
  else:
    print(ch)
```
