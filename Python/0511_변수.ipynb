##인덱스 & 인덱싱
- Python은 제로 인덱스 : 0부터 시작<br>
- e.g) hello -> 0번 인덱스는 h, 4번 인덱스는 o

```python
print("안녕하세요"[0])
print("안녕하세요"[0:3]) #0,1,2 인덱스 출력
print("안녕하세요"[0:-1])
print("안녕하세요"[:-1:2])
print("안녕하세요"[:-1])
print("abcdefghijklmnop"[::3])
print("안녕하세요"[-1:])
print("안녕하세요"[1:1+2])
```

```python
#문자열의 길이
print(len("hello"))

a=len("안녕하세요.")
print(a)

print(len("hello"[1:4]))
```

##복합 대입 연산자
1) += 더하기 후에 대입<br>

2) -= 빼기 후 대입<br>

4) /= 나누기 후 대입 <br>
<br>
그 외 //,%,** 모두 같음

```python
a = 10
a += 1 #a=10+1=11
a += 1 #a=11+1=12
a += 1 #a=12+1=13
print(a)
```

```python
a = 10
a += 1 #11
a -= 2 #9
a *= 2 #18
a //= 4 #4
print(a)
```

##실습 : m -> cm 변환 프로그램

```python
# 1m = 100cm
meter = input("키(m)를 입력하세요 > ")
meter = float(meter)
height = meter * 100
print(meter, "m 는 cm 로 변환하면 ", height, "cm 입니다.")
```

### inch -> cm 변환 프로그램

```python
#1 inch = 2.54 cm
inch = input("길이(inch)를 입력하세요 > ")
inch = float(inch)
length= inch * 2.54
print(inch, "inch 는 cm 로 변환하면 ", length, "cm 입니다.")
```
