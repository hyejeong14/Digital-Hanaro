```python
import pandas as pd
# 데이터 불러오기
# csv 파일 불러오기 + 열 이름이 없음(header = None)
df = pd.read_csv('/content/drive/MyDrive/Digital_Hana1/data/auto-mpg.csv', header = None)
```

```python
# 열 이름 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
              'acceleration', 'model_year', 'origin', 'car_name']
```

```python
# 연비 mpg = mile per gallon
# kpl = kilometer per liter
# 1 mpg = 0.425 kpl
kpl = 0.425
df['kpl'] = df['mpg'] * kpl
df['kpl'] = df['kpl'].round(1)
print(df.head())
```

##구간 분할

```python
#마력 실수형 => 수치형 변수 => 범주형 변수
print(df.horsepower.dtypes)
print(df.horsepower[:5])
```

```python
#구간 설정
#1) 선행 연구 등 참고
#2) 데이터 기반 정의 = 예) 히스토그램 구간 설정
#히스토그램을 이용해 구간과 빈도 추출
count, bins = np.histogram(df['horsepower'],bins=3)
```

```python
print(count,bins)
```

```python
bins_name = ['저출력','중간출력','고출력']
df['hp_cat'] = pd.cut(x=df['horsepower'],  #사용할 데이터
                      bins = bins,  #구간의 값
                      labels = bins_name, #구간의 이름
                      include_lowest=True) #구간의 시작값 포함 설정
```

```python
print(bins,'\n',df['hp_cat'])
```

##정규화

```python
#최대절대값 정규화
#관측치/절대값(최대값) 
#=> 절대값의 씌웠을 때의 최대값을 구해야함 ! 아니면 음수 양수 섞였을 때 문제 발생 *
#관측치 = -최대값 => -1
#관측지 = 최대값 => 1
#-1~1 사이의 값을 가짐
df['hp_nmz1'] = df['horsepower']/abs(df['horsepower']).max()
df['hp_nmz1'].describe()
```

```python
#최대-최소 정규화
#(관측치-최소값)/(최대값-최소값)
#관측치 = 최소값 => 0
#관측치 = 최대값 => 1
#0~1 사이의 값을 가짐
df['hp_nmz2'] = (df['horsepower']-df['horsepower'].min())/(df['horsepower'].max()-df['horsepower'].min())
df['hp_nmz2'].describe()
```

###시리즈 객체에 적용

```python
#1) 반환하는 값이 여러 개인 경우
#null값에 대한 bool 인덱스를 반환하는 함수
def mv_find(series):
  return series.isnull()
```

```python
df.apply(mv_find,axis=0) #열 단위 = 시리즈 객체
```

```python
#2) 반환하는 값이 하나인 경우
#범위를 구하는 함수
def range_xy(series):
  return series.max() - series.min()
```

```python
df.apply(range_xy, axis = 0)
```

```python
#행 단위
df['age_plus_five']=df.apply(lambda x : x['age']+x['five'], axis=1)
```

```python
df.apply(lambda x : add_two(x['age'],x['five']),axis=1)
```

###데이터프레임 객체에 적용

```python
df=df.loc[:,'age':'fare']
```

```python
df1 = df.pipe(mv_find)
```

```python
type(df1) #데이터프레임 = 함수가 적용된 결과는 시리즈 개수와 일치(891개)
```

```python
df1
```

```python
def mv_count(series):
  return mv_find(series).sum()
```

```python
df2 = df.pipe(mv_count)
print(type(df2));print(df2) #시리즈 = 함수 적용된 결과가 하나 (열 마다)
```

```python
def mv_count_total(series):
  return mv_count(series).sum()
```

```python
df3 = df.pipe(mv_count_total) 
print(type(df3));print(df3) #결과가 하나의 개별값으로 출력 = 함수 적용된 결과가 개별값
```
