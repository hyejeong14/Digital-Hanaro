##데이터프레임 합치기

###연결

```python
# 위아래로 합치기 concat(axis = 0)
# 좌우로 합치기 concat(axis = 1)
import pandas as pd
df1 = pd.DataFrame({'a': [1,2,3,4],
                    'b': [11,12,13,14],
                    'c': [21,22,23,24]})
df2 = pd.DataFrame({'a': [9,8,7],
                   'b': [19,18,17],
                   'c': [29,28,27],
                   'd': [39,38,37]})
print(df1)
print(df2)
```

```python
#기본값 => axis=0, join='outer'
result_1 = pd.concat([df1,df2])
```

```python
result_2 = pd.concat([df1,df2],join = 'inner')
result_2
```

```python
result_3 = pd.concat([df1,df2],join = 'inner',ignore_index=True)
result_3
```

```python
#좌우로 합치기 concat(axis=1)
result_4 = pd.concat([df1,df2],axis=1,join='outer')
result_4
```

```python
result_5 = pd.concat([df1,df2],axis=1,join='inner')
result_5
```

```python
sr1 = pd.Series([4,5,6,7],name='e')
sr2 = pd.Series([14,15,16,17],name='f',index=[1,2,3,4])
sr3 = pd.Series([24,25,26,27],name='g')
```

```python
print(sr1);print(sr2);print(sr3)
```

```python
#시리즈를 axis=1로 합치면 새로운 열을 합치는 것과 동일
#시리즈의 이름이 열 이름
print(df1)
pd.concat([df1,sr1],axis=1)
```

```python
#join='outer' 행 인덱스가 같은 것끼리 연결
#없는 인덱스인 경우 NaN으로 표현
print(df2)
pd.concat([df2,sr2],axis=1)
```

```python
#공통된 행만 출력
print(df2)
pd.concat([df2,sr2],axis=1,join='inner')
```

###병합

```python
#merge = SQL JOIN과 유사 => 좌우로 병합
#공통된 열, 인덱스가 존재해야 함 => 키
import pandas as pd
```

```python
df1 = pd.read_excel('/content/drive/MyDrive/Digital_Hana1/data/stock price.xlsx',engine='openpyxl')
df2 = pd.read_excel('/content/drive/MyDrive/Digital_Hana1/data/stock valuation.xlsx',engine='openpyxl')

print(df1.head());print(df2.head())
```

```python
#기본값 how = 'inner' , on = None() 두 데이터프레임이 공통으로 가지고 있는 열 사용
pd.merge(df1,df2)
```

```python
#FULL OUTER JOIN 
merge_outer= pd.merge(df1,df2,how='outer',on='id')
merge_outer.head()
```

```python
#LEFT OUTER JOIN 
merge_left= pd.merge(df1,df2,how='left',on='id')
merge_left.head()
```

```python
#RIGHT OUTER JOIN 
merge_right= pd.merge(df1,df2,how='right',on='id')
merge_right.head()
```

```python
#불 인덱스
df1_price = df1[df1['price']<100000]
```

```python
pd.merge(df1_price, df2)
```

###결합

```python
#join = merge 와 유사
#차이점 => join은 행 인덱스 기준으로 결합하는 방식
#옵션 지정시 열 기준으로 결합 가능
print(df1);print(df2)
```

```python
#RIGHT OUTER JOIN
join_right = df2.join(df1)
join_right
```

```python
#INNER JOIN
join_inner = df1.join(df2,how='inner')
join_inner
```

###그룹 연산 메소드

```python
#클래스 그룹 객체 생성
grouped = df.groupby('class')
```

```python
#표준 편차 => 내장함수 사용
gr_sd = grouped .std()
```

```python
print(gr_sd)
print(type(gr_sd))
```

```python
gr_age_sd = grouped.age.std()
print(gr_age_sd)
print(type(gr_age_sd)) #시리즈 형태 = 하나의 열을 가짐
```

```python
#agg
#사용자 정의 함수 생성 ( 범위 반환 )
def range_xy(x):
  return x.max()-x.min()
```

```python
#그룹 객체에 사용자 정의 함수 적용
grouped.agg(range_xy)
```

```python
grouped.agg(['min','max'])
```

```python
grouped.agg({'age':['min','max'],'fare':'mean'})
```

```python
#transform = 그룹 연산 데이터 변환
#단일행 함수 = 데이터 개수와 함수 적용 결과 개수가 동일
age_mean = grouped.age.mean()
age_sd = grouped.age.std()
print(age_mean,'\n');print(age_sd)
```

```python
for i,j in grouped.age:
  z_score = (j-age_mean[i])/age_sd[i]
  print('그룹 이름 : ',i)
  print(z_score,'\n')
```

```python
def zscore(x):
  return (x - x.mean())/x.std()
```

```python
grouped.age.transform(zscore) #전체 데이터 개수 891개 = First 개수 + Second 개수 + Third 개수
```

```python
#계산 가능한 모든 열에 대해서 적용 가능
grouped.transform(zscore)
#survived는 범주형 변수이기 때문에 의미x
```

##피봇테이블

```python
pd.pivot_table(df,   #데이터
               index = 'class',  #행
               columns = 'sex',  #열
               values = 'age', #값
               aggfunc = 'mean') #집계 함수 설정
```

```python
pd.pivot_table(df,
               index = 'class',
               columns = 'sex',
               values = 'age',
               aggfunc = ['mean','std']) #함수 2개 사용
```

```python
pd.pivot_table(df,
               index = ['class','sex'], # 행 2개 사용
               columns = 'survived',
               values = 'age',
               aggfunc = ['mean','std']) #함수 2개 사용
```

```python
pv_table = pd.pivot_table(df,
                          index = ['class','sex'],
                          columns = 'survived',
                          values = ['age','fare'],
                          aggfunc = ['mean','std']) #함수 2개 사용'

pv_table
```

```python
#First 클래스만 보기
pv_table.xs('First')
```

```python
#First 클래스, female만 보기 (시리즈 형태)
pv_table.xs(('First','female'))
```

```python
# 데이터프레임 형태로 보기
pv_table.xs(('First','female'),level=['class','sex'])
```

```python
#평균만 보기
pv_table.xs('mean', axis=1)
```

```python
#평균만 보기
pv_table.xs(('mean','age',1), level = [0,1,2], axis=1)
```
